# Tata IPL Auction Stats 2023 - Web Scaping

```python
# import libraries

import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
## sending request to the server to read the url using Requests and BeautifulSoup lib 
# fetching the html content for website

url = 'https://www.iplt20.com/auction'
r = requests.get(url)
print(r)
```

```python
# find all the table data

soup = BeautifulSoup(r.text,'lxml')
print(soup)
```

```python
# find all the table data

table = soup.find('table',class_='ih-td-tab auction-tbl')
print(table)
```

```python
# collect all the text from the data
title = table.find_all('th')
print(title)
```

```python
## fetching required header from the table data

header = []
for i in title:
    name = i.text
    header.append(name)

print(header)
```

```python
# convert list into DataFrame
df = pd.DataFrame(columns=header)
df_temp = df.drop_duplicates()
print(df)
```

```python
## find all the rows data from table 

rows = table.find_all('tr')
print(rows)
```

```python
## first collect all the required rows from table
# then append rows data in the header
# then append all the data in df_temp dataframe

for i in rows[1:]:
    first_td = i.find_all('td')[0].find('div', class_='ih-pt-ic').text.strip()
    data = i.find_all('td')[1:]
  #  print(data)
    row = [tr.text for tr in data]
  #  print(row)
    row.insert(0,first_td)
    l = len(df_temp)
    df_temp.loc[l] = row
print(df_temp)
```

```python
# Export dataframe into csv file
df_temp.to_csv('IPL Auction Stats.csv',index = False)
```
