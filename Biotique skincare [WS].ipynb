```python
# import libraries

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# sending request to the website for parsing html using requests libraries
url = 'https://www.biotique.com/collections/skincare'
r = requests.get(url)
soup = BeautifulSoup(r.text,'lxml')
print(soup)
```

```python
## passing product list name
Product_Name = []
Description = []
Price = []
```

```python
# scrap product name from html content
names = soup.find_all('span',class_='title')
print(names)
```

```python
# scrap product desc from html content
desc = soup.find_all('div',class_='custom-features')
print(desc)
```

```python
# scrap product price from html content
price = soup.find_all('span',class_='money')
print(price)
```

```python
## append data in original list
names = soup.find_all('span',class_='title')
for i in names:
    n = i.text
    Product_Name.append(n)

desc = soup.find_all('div',class_='custom-features')
for i in desc:
    n = i.text.strip('\n')
    Description.append(n)
    
price = soup.find_all('span',class_='money')
for i in price:
    n = i.text
    Price.append(n)
```

```python
# convert the data into dataframe
df = pd.DataFrame({'Product Name':Product_Name,'Description':Description,'Price':Price})
df_temp = df.drop_duplicates()
#print(df_temp)
```

```python
print(df_temp)
```

```python
# Export dataframe into csv file
df_temp.to_csv('Biotique Skincare Dataset.csv',index=False)
```
